```python
#bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='CYLINDERS', y='CO2EMISSIONS', data=engine_df)
plt.title('Bar chart of CO2 emissions vs Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
#Scatter plot
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='ENGINESIZE', y='CO2EMISSIONS', data= engine_df)
plt.title('Scattor plot of Co2 Emission vs Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Co2 Emission')
plt.show()

    
```

```python
#Histogram
```

```python
plt.figure(figsize=(8,4))
sns.histplot(engine_df['FUELCONSUMPTION_COMB'], bins=20, kde=True)
plt.title('Histogram fuel consumption (combined)')
plt.xlabel('Fuel comsumption (combined)')
plt.ylabel('Frequency')
plt.show()
```

```python
#line plot
plt.figure(figsize=(8,4))
sns.lineplot(x='ENGINESIZE', y='FUELCONSUMPTION_COMB', data=engine_df)
plt.title('Line plot of fuel consumption VS Engine Size')
plt.xlabel('Engine size')
plt.ylabel('fuel consumption(combine)')
plt.show()
```
