```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt_df = pd.read_csv('/home/hp/DBDA/ML/machine-learning/Assignments/delivery_time.csv')
```

```python
print("\n 1st 5 record of data")
print(dt_df.head(5))
print("="*100)
```

```python
print("\n last 5 record of data")
print(dt_df.tail(5))
print("=" * 100)
```

```python
print("\n info and describe the data")
print(dt_df.info())
print(dt_df.describe())
print("="* 100)
```

```python
print("\n null values of data")
print(dt_df.isnull().sum())
print("="* 100)
```

```python
print(dt_df.columns)
```

```python
#bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Delivery Time', y='Sorting Time', data=dt_df)
plt.title('Delivery time VS Sorting Time')
plt.xlabel('Delivery Time')
plt.ylabel('Sorting Time')
plt.show()
```

```python
#Scattor plot
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='Delivery Time', y='Sorting Time', data= dt_df)
plt.title('Scattor plot of Delivery Time VS Sorting Time')
plt.xlabel('Delivery Time')
plt.ylabel('Sorting Time')
plt.show()
```

```python
#histogram
```

```python
plt.figure(figsize=(8,4))
sns.histplot(dt_df['Delivery Time'], bins=20, kde=True)
plt.title('Delivery Time And Sorting time')
plt.xlabel('Delivery Time')
plt.ylabel('Sorting Time')
plt.show()
```

```python
#line plot
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x='Delivery Time', y='Sorting Time', data=dt_df)
plt.title('line plot of Delivery time and Sorting Time')
plt.xlabel('Delivery time')
plt.ylabel('Sorting Time')
plt.show()
```

```python
#pair plots
```

```python
sns.pairplot(dt_df[['Delivery Time','Sorting Time']])
plt.show()
```
