```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tv_df = pd.read_csv('/home/hp/DBDA/ML/machine-learning/Assignments/TvMarketing.csv')
```

```python
print("\n first five records")
print(tv_df.head(5))
print("=" * 100)
```

```python
print("\n last five records")
print(tv_df.tail(5))
print("=" * 100)
```

```python
print("\n info and describe the data")
print(tv_df.info())
print(tv_df.describe())
print("="* 100)
```

```python
print("\n null values of data")
print(tv_df.isnull().sum())
print("="* 100)
```

```python
print(tv_df.columns)
```

```python
#bar plot
plt.bar(tv_df["TV"],tv_df["Sales"])
plt.title('TV and Sales')
plt.xlabel('Tv')
plt.ylabel('Sales')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TV', y='Sales', data=tv_df)
plt.title('Tv vs Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='TV', y='Sales', data= tv_df)
plt.title('Scattor plot of TV and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')

plt.show()
```

```python
#histogram
plt.figure(figsize=(8,4))
sns.histplot(tv_df['TV'], bins=20, kde=True)
plt.title('Tv and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
#line plot
plt.figure(figsize=(8,4))
sns.lineplot(x='TV', y='Sales', data=tv_df)
plt.title('line plot of Tv and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')

plt.show()
```

```python
sns.pairplot(tv_df[['TV', 'Sales']])
plt.show()
```
