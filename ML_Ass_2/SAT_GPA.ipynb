```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
st_df = pd.read_csv('/home/hp/DBDA/ML/machine-learning/Assignments/SAT-GPA.csv')
```

```python
#bar chart
import numpy as np
plt.bar(st_df["SAT"],st_df["GPA"])
plt.title('SAT VS GPA')
plt.xlabel('SAT')
plt.ylabel('GPA')
```

```python
st_df["SAT"]
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='SAT', y='GPA', data= st_df)
plt.title('Scattor plot of SAT VS GPA')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.legend()
plt.show()
```

```python
#histogram
```

```python
plt.figure(figsize=(8,4))
sns.histplot(st_df['SAT'], bins=20, kde=True)
plt.title('SAT and GPA')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
#line plot
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x='SAT', y='GPA', data=st_df)
plt.title('line plot of SAT and GPA')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.legend()
plt.show()
```
