```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/hp/DBDA/ML/machine-learning/Assignments/claimants.csv")
```

```python
df.head(10)
```

```python
df.isna().any()
```

```python
df["CLMSEX"].fillna(df['CLMSEX'].mode()[0],inplace=True)

df['CLMAGE'].fillna(df['CLMAGE'].mean(),inplace=True)

df["SEATBELT"].fillna(df['SEATBELT'].mode()[0],inplace=True)
df["CLMINSUR"].fillna(df["CLMINSUR"].mode()[0],inplace=True)
```

```python
df.corr()
```

```python
x = df[['CLMAGE', 'CLMSEX', 'SEATBELT', 'CLMINSUR', 'LOSS']]
y = df['ATTORNEY']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=123456)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)
```

```python
from sklearn.svm import SVC
model = SVC(C=3, gamma="auto")
model.fit(x_train, y_train)
```

```python
def function(model, model_name):
    pred = model.predict(x_test)
    print(metrics.classification_report(pred, y_test), model_name)
    print(metrics.confusion_matrix(pred, y_test))
```

```python
function(model, "naive_bayes")
```

```python
function(model, "KNeighborsClassifier")
```
